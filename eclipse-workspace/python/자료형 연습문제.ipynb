```python
# 문제 8
# 주어진 리스트 [1, 2, 3, 4, 5]에서 첫 번째 원소와 마지막 원소를 교환하는 파이썬 코드를 작성하세요.
#자바, 파이썬 둘 다 가능한 방식
a = [1, 2, 3, 4, 5]
temp = a[0]
a[0] = a[4]
a[4] = temp
print(a)
# 파이썬에서만 가능한 방식
a = [1, 2, 3, 4, 5]
a[0],a[4] = a[4],a[0]
print(a)
```
