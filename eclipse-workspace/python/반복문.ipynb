```python
# while문
a = 0 # 초기값
while(a < 10): # 조건식
    a += 1 # 증감식
    print("실행문 :", a)
```

```python
# break, continue문
a = 0 # 초기값
while(True): # 조건식
    a += 1 # 증감식
    if(a > 5):
        break # 반복문 강제 종료
    if(a % 2 == 0):
        continue # 다음 반복으로 건너뛰기
    print("실행문 :", a)
```

```python
# for문
# for 변수 in 반복가능한자료형:
a = [5,4,3,2,1]
for i in a:
    print(i)
b = ["홍길동", "이순신", "손오공"]
for i in b:
    print(i)
```

```python
# for문과 range함수
print(list(range(5))) # 0 ~ 4 까지
print(list(range(0, 10+1, 2))) # 0,2,4,6,8,10
# range(시작값, 끝값 + 1)
for i in range(1,10+1):
    print(i)
```
